```python
#Problem 8:. Email Management Create a class representing an email with attributes like sender, recipient, and subject. Implement methods to send an email and display email details
import smtplib
from email.message import EmailMessage
from email.utils import formatdate

class Email:
    def __init__(self, sender, recipient, subject):
        self.sender = sender
        self.recipient = recipient
        addedsubject = subject
        self.subject = addedsubject
        self.content = ""
        self.received_at = None
        self.password = 'PASSWORD'

    def send(self):
        """Sends the email to the recipient"""
        # Generate the message
        msg = EmailMessage()
        msg.set_content(self.content)
        msg['Subject'] = self.subject
        msg['From'] ,'From: {self.sender}'.format(self=self)
        msg['To'] ,'To: {self.recipient}'.format(self=self)
        msg['Date'] ,formatdate(localtime=True)

        # Send the message
        try:
            server = smtplib.SMTP('smtp.gmail.com', 587)
            server.starttls()
            server.login(self.sender, self.password)
            server.send_message(msg)
            server.quit()
            self.received_at = formatdate.now()
        except:
            print("Error: Unable to send email")

    def display_details(self):
        """Displays the email details in a readable format"""
        print(f"From: {self.sender}")
        print(f"To: {self.recipient}")
        print(f"Subject: {self.subject}")
        print(f"Received at: {self.received_at}")
        print(f"Content: {self.content}")

# Example usage
email = Email("example@gmail.com", "rexample@gmail.com", "Important Email")
email.content = "This is the content of the email."
email.send()
email.display_details()
```

```python
#Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.
class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task, due_date):
        self.tasks.append((task, due_date, False))

    def mark_task_as_completed(self, task_name):
        for task in self.tasks:
            if task[0] == task_name:
                task = (task[0], task[1], True)
                break

    def display_pending_tasks(self):
        for task in self.tasks:
            if task[2] == False:
                print(task[0], '(', task[1], ')')
```
